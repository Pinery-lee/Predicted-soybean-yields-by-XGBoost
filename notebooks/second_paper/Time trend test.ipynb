```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# soybean
# import data
data = pd.read_csv('D:\论文-产量趋势利用\数据\input_soybean_exp.csv')
# D:\论文-产量趋势利用\数据\input_soybean.csv
```

```python
# 全中西部
linear_m = LinearRegression()
metric_list = []
data_train_val1 = data[data['Year'].isin([i for i in range(1980,2022)])]
X99 = np.array(data_train_val1['Year']).reshape(-1,1)
Y99 = np.array(data_train_val1['yield(t/ha)']).reshape(-1,1)
linear_m.fit(X99, Y99)

X99 = sm.add_constant(X99) # 向 x1 左侧添加截距列 x0=[1,...1]
model = sm.OLS(Y99, X99)
results = model.fit()
# print(results.summary())
print('The foluma of linea-trend the yield: yield = {:.3f} × year + ({:.3f})'
      .format(linear_m.coef_[0,0],linear_m.intercept_[0]))

print('Number:',len(X99))
print('R-squared:',round(results.rsquared,3))
print('P-value:',round(results.pvalues[-1],3))
metric_list.append([ len(X99), linear_m.coef_[0,0], linear_m.intercept_[0], round(results.rsquared,3),round(results.pvalues[-1],3)])
name = ['Number', 'slope', 'bias', 'R-squared', 'P-value']
metric_df = pd.DataFrame(columns = name, data=metric_list)
metric_df.to_csv('D:/论文-产量趋势利用/结果/趋势检测/soybean/' +'大豆中西部线性趋势检测结果.csv',header=True)
```

```python
# 分州统计
metric_list = []
linear_m = LinearRegression()
for state in set(data.State):
    data1 = data[data['State'].isin([state])]
    data_train_val1 = data1[data1['Year'].isin([i for i in range(1980,2022)])]

    X99 = np.array(data_train_val1['Year']).reshape(-1,1)
    Y99 = np.array(data_train_val1['yield(t/ha)']).reshape(-1,1)
    linear_m.fit(X99, Y99)

    X99 = sm.add_constant(X99) # 向 x1 左侧添加截距列 x0=[1,...1]
    model = sm.OLS(Y99, X99)
    results = model.fit()
    print('The foluma of linea-trend the yield: yield = {:.3f} × year + ({:.3f})'
          .format(linear_m.coef_[0,0],linear_m.intercept_[0]))
    
    print('Number:',len(X99))
    print('state:',state)
    print('R-squared:',round(results.rsquared,3))
    print('P-value:',round(results.pvalues[-1],3))
    metric_list.append([state, len(X99), linear_m.coef_[0,0], linear_m.intercept_[0], round(results.rsquared,3),round(results.pvalues[-1],3)])
    name = ['state', 'Number', 'slope', 'bias', 'R-squared', 'P-value']
    metric_df = pd.DataFrame(columns = name, data=metric_list)
    metric_df.to_csv('D:/论文-产量趋势利用/结果/趋势检测/soybean/' +'大豆州级线性趋势检测结果.csv',header=True)
```

```python
# 分县统计
linear_m = LinearRegression()
metric_list = []
for sta_con in set(data.sta_con):
    data1 = data[data['sta_con'].isin([sta_con])]
    data_train_val1 = data1[data1['Year'].isin([i for i in range(1980,2022)])]

    X99 = np.array(data_train_val1['Year']).reshape(-1,1)
    Y99 = np.array(data_train_val1['yield(t/ha)']).reshape(-1,1)
    linear_m.fit(X99, Y99)

    X99 = sm.add_constant(X99) # 向 x1 左侧添加截距列 x0=[1,...1]
    model = sm.OLS(Y99, X99)
    results = model.fit()
    # print(results.summary())
    print('The foluma of linea-trend the yield: yield = {:.3f} × year + ({:.3f})'
          .format(linear_m.coef_[0,0],linear_m.intercept_[0]))
    
    print('Number:',len(X99))
    print('sta_con:',sta_con)
    print('R-squared:',round(results.rsquared,3))
    print('P-value:',round(results.pvalues[-1],3))
    metric_list.append([sta_con, len(X99), linear_m.coef_[0,0], linear_m.intercept_[0], round(results.rsquared,3),round(results.pvalues[-1],3)])
    name = ['county', 'Number', 'slope', 'bias', 'R-squared', 'P-value']
    metric_df = pd.DataFrame(columns = name, data=metric_list)
    metric_df.to_csv('D:/论文-产量趋势利用/结果/趋势检测/soybean/' +'大豆县级线性趋势检测结果.csv',header=True)
```
