```python
import pandas as pd
def read_data(path):
    '''读取数据'''
    try:
        data = pd.read_csv(path, engine='python', encoding='utf8')
        return data
    except FileNotFoundError:
        print('File not found. Please check the path and filename.')
        return None
    except Exception as e:
        print('An error occurred:', e)
        return None
```

```python
# 大豆列名重命名
# 2014-2021，
data = read_data('D:/毕业大论文/数据/input_soybean.csv')
# 时间变量
# 创建包含需要替换的列名和新列名的字典
column_dict = {
    '7_NDVI': 'NDVI_Mid Mar.',
    '7_EVI': 'EVI_Mid Mar.',
    '7_LSWI': 'LSWI_Mid Mar.',
    '7_GCVI': 'GCVI_Mid Mar.',
    '7_RVI': 'RVI_Mid Mar.',
    '7_SAVI': 'SAVI_Mid Mar.',
    '7_WDRVI': 'WDRVI_Mid Mar.',
    '7_Fpar': 'Fpar_Mid Mar.',
    '7_Lai': 'LAI_Mid Mar.',
    '7_ET': 'ET_Mid Mar.',
    '7_LE': 'LE_Mid Mar.',
    '7_LST_Day_1km': 'LST_Day_Mid Mar.',
    '7_LST_Night_1km': 'LST_Night_Mid Mar.',
    '7_spi14d': 'SPI14d_Mid Mar.',
    '7_spi30d': 'SPI30d_Mid Mar.',
    '7_spi90d': 'SPI90d_Mid Mar.',
    '7_eddi14d': 'EDDI14d_Mid Mar.',
    '7_eddi30d': 'EDDI30d_Mid Mar.',
    '7_eddi90d': 'EDDI90d_Mid Mar.',
    '7_spei14d': 'SPEI14d_Mid Mar.',
    '7_spei30d': 'SPEI30d_Mid Mar.',
    '7_spei90d': 'SPEI90d_Mid Mar.',
    '7_pdsi': 'PDSI_Mid Mar.',
    '7_z': 'Palmer Z_Mid Mar.',
    '7_sur_refl_b01': 'Red_Mid Mar.',
    '7_sur_refl_b02': 'Nir_Mid Mar.',
    '7_sur_refl_b03': 'Blue_Mid Mar.',
    '7_sur_refl_b04': 'Green_Mid Mar.',
    '7_sur_refl_b05': 'Nir1_Mid Mar.',
    '7_sur_refl_b06': 'Swir1_Mid Mar.',
    '7_sur_refl_b07': 'Swir2_Mid Mar.',
    '7_total_precipitation': 'Precipitation_Mid Mar.',
    '7_temperature': 'Temp_Mid Mar.',
    '7_specific_humidity': 'Humidity_Mid Mar.',
    '7_pressure': 'Pressure_Mid Mar.',
    '7_shortwave_radiation': 'Shortwave_Mid Mar.',
    '7_longwave_radiation': 'Longwave_Mid Mar.'
}
column_new = column_dict
# 对于每个时间段重复相同的步骤
# 例如，对于8月份：
for number, month in {
    '8_':'Late Mar.',
    '9_':'Early Apr.',
    '10_':'Mid Apr.',
    '11_':'Late Apr.',
    '12_':'Early May.',
    '13_':'Mid May.',
    '14_':'Late May.',
    '15_':'Early Jun.',
    '16_':'Mid Jun.',
    '17_':'Late Jun.',
    '18_':'Early Jul.',
    '19_':'Mid Jul.',
    '20_':'Late Jul.',
    '21_':'Early Aug.',
    '22_':'Mid Aug.',# 八月中旬
    '23_':'Late Aug.',
    '24_':'Early Sep.',
    '25_':'Mid Sep.',
    '26_':'Late Sep.',
    '27_':'Early Oct.',
    '28_':'Mid Oct.',
    '29_':'Late Oct.',
    '30_':'Early Nov.',
    '31_':'Mid Nov.',
    '32_':'Late Nov.',
    '33_':'Early Dec.'}.items():
    
    new_dict = {k.replace('7_', number): v.replace('Mid Mar.', month) for k, v in column_dict.items() if k.startswith('7_')}
    column_new.update(new_dict)
# 物候变量
# 创建包含需要替换的列名和新列名的字典
column_dict1 = {
    'Planted_NDVI':'NDVI_Planting', # Planted
    'Planted_EVI':'EVI_Planting', 
    'Planted_LSWI':'LSWI_Planting', 
    'Planted_GCVI':'GCVI_Planting', 
    'Planted_RVI':'RVI_Planting', 
    'Planted_SAVI':'SAVI_Planting', 
    'Planted_WDRVI':'WDRVI_Planting', 
    'Planted_Fpar':'Fpar_Planting', 
    'Planted_Lai':'LAI_Planting', 
    'Planted_LE':'LE_Planting', 
    'Planted_LST_Day_1km':'LST_Day_Planting', 
    'Planted_LST_Night_1km':'LST_Night_Planting', 
    'Planted_spi14d':'SPI14d_Planting', 
    'Planted_spi30d':'SPI30d_Planting', 
    'Planted_spi90d':'SPI90d_Planting', 
    'Planted_eddi14d':'EDDI14d_Planting', 
    'Planted_eddi30d':'EDDI30d_Planting', 
    'Planted_eddi90d':'EDDI90d_Planting', 
    'Planted_spei14d':'SPEI14d_Planting', 
    'Planted_spei30d':'SPEI30d_Planting', 
    'Planted_spei90d':'SPEI90d_Planting', 
    'Planted_pdsi':'PDSI_Planting', 
    'Planted_z':'Palmer Z_Planting', 
    'Planted_sur_refl_b01':'Red_Planting', 
    'Planted_sur_refl_b02':'Nir_Planting', 
    'Planted_sur_refl_b03':'Blue_Planting', 
    'Planted_sur_refl_b04':'Green_Planting', 
    'Planted_sur_refl_b05':'Nir1_Planting', 
    'Planted_sur_refl_b06':'Swir1_Planting', 
    'Planted_sur_refl_b07':'Swir2_Planting',
    'Planted_temperature':'Temp_Planting', 
    'Planted_shortwave_radiation':'Shortwave_Planting', 
    'Planted_longwave_radiation':'Longwave_Planting', 
    'Planted_ET':'ET_Planting'}
column_new.update(column_dict1)

# 对于每个物候阶段重复相同的步骤
# 例如，对于emerged月份：
for phology_k, phology_v in {
    'Emergrd_':'_Emerging',
    'Blooming_':'_Blooming',
    'Setting Pods_':'_Podding',
    'Dropping Leaves_':'_Dropping leaves',
    'Harvested_':'_Harvest',}.items():
    new_dict1 = {k.replace('Planted_', phology_k): v.replace('_Planting', phology_v) for k, v in column_dict1.items() if k.startswith('Planted_')}
    column_new.update(new_dict1)

# 对于静态变量
column_dict2 = {
    'irrigated soybean area':'AREA_Irrigated',
    'soybean area':'AREA',
    'PIC':'PIC',
    'caco3_kg_sq_m':'CaCO3',
    'cec_025':'CEC',
    'drainage_class_int':'Drainage', 
    'ec_025':'EC',
    'i_class':'I_class',
    'n_class':'N_class', 
    'max_om':'Max_OM', 
    'paws_025':'PAWS', 
    'ph_025':'pH', 
    'sar':'SAR', 
    'texture_025':'Texture', 
    'sand_025':'Sand', 
    'silt_025':'Silt', 
    'clay_025':'Clay'}
column_new.update(column_dict2)

# 重命名
data = data.rename(columns=column_new)
data.to_csv('D:/毕业大论文/数据/input_soybean_rename.csv', index=False)
print(data.columns.tolist())
```

```python
# 大豆州one-hot
# Encoding categry feature
# one-hot encoding
data = read_data('D:/毕业大论文/数据/input_soybean_rename.csv')
onehot_cols = ['State']
new_onehot_cols = pd.get_dummies(data[onehot_cols]).columns.tolist()
data.loc[:,new_onehot_cols] = pd.get_dummies(data[onehot_cols])
data = data.rename(columns = {
    'State_kansas': 'KS',
    'State_north dakota': 'ND',
    'State_michigan': 'MI',
    'State_minnesota': 'MN',
    'State_wisconsin': 'WI',
    'State_ohio': 'OH',
    'State_iowa': 'IA',
    'State_missouri': 'MO',
    'State_south dakota': 'SD',
    'State_indiana': 'IN',
    'State_illinois': 'IL',
    'State_nebraska': 'NE'})
data.to_csv('D:/毕业大论文/数据/input_soybean_rename.csv', index=False)
```
